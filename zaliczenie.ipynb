```python
Zad. 1

% Fakty
ojciec(jan, anna).
ojciec(jan, piotr).
matka(maria, anna).
matka(maria, piotr).

ojciec(grzegorz, jan).
matka(ala, jan).
ojciec(grzegorz, patryk).
matka(ala, patryk).
ojciec(patryk, kevin).
matka(kasia, kevin).
mezczyzna(jan).
kobieta(maria).
mezczyzna(piotr).
kobieta(anna).
mezczyzna(grzegorz).
mezczyzna(patryk).
mezczyzna(kevin).
kobieta(ala).
kobieta(kasia).
malzenstwo(jan, maria).
malzenstwo(grzegorz, ala).
malzenstwo(patryk, kasia).

% Reguły
rodzic(X, Y) :- ojciec(X, Y); matka(X, Y).

przodek(X, Y) :- rodzic(X, Y).

brat(X, Y) :- mezczyzna(X), rodzic(Z, X), rodzic(Z, Y), X \= Y.

siostra(X, Y) :- kobieta(X), rodzic(Z, X), rodzic(Z, Y), X \= Y.
kuzyn(X, Y) :- mezczyzna(X), rodzic(Z1, X), rodzic(Z2, Y), brat(Z1, Z2).
kuzyn(X, Y) :- mezczyzna(X), rodzic(Z1, X), rodzic(Z2, Y), siostra(Z1, Z2).

?- przodek(jan, anna). % Oczekiwane: true
?- brat(piotr, anna). % Oczekiwane: true
?- kuzyn(X, piotr). % Oczekiwane: X = Kevin


Zad.2

% Fakty
lotnisko(warszawa).
lotnisko(krakow).
lotnisko(gdansk).

polaczenie(warszawa, krakow, 300).
polaczenie(krakow, gdansk, 200).
polaczenie(warszawa, gdansk, 500).
polaczenie(warszawa, krakow, 300).
polaczenie(krakow, gdansk, 200).
polaczenie(warszawa, gdansk, 500).
polaczenie(warszawa, lodz, 120).
polaczenie(lodz, krakow, 180).
polaczenie(krakow, gdansk, 200).
polaczenie(warszawa, bialystok, 190).
polaczenie(bialystok, olsztyn, 150).
polaczenie(olsztyn, gdansk, 120).
polaczenie(warszawa, poznan, 300).
polaczenie(poznan, wroclaw, 160).
polaczenie(wroclaw, krakow, 270).
polaczenie(warszawa, gdansk, 500).


% Reguły
polaczenie(X, Y) :- polaczenie(X, Y, _). % Zakładamy, że połączenia są dwustronne.

koszt_podrozy(X, Y, C) :- polaczenie(X, Y, C).

droga(A, B, [A, B], X) :-
    polaczenie(A, B, X).
droga(A, B, PathAB, Length) :-
    polaczenie(A, C, X),
    droga(C, B, PathCB, LengthCB),
    PathAB = [A | PathCB],
    Length is X + LengthCB.

    ?- polaczenie(warszawa, krakow).

    Oczekiwany wynik: true

    ?- koszt_podrozy(warszawa, gdansk, C).

    Oczekiwany wynik: C = 500

Zad. 3
% Fakty
produkt(tv, 10, sektor_a).
produkt(pralka, 5, sektor_b).
produkt(lodowka, 3, sektor_c).

% Reguły
dostepny_produkt(X) :- produkt(X, Ilosc, _), Ilosc > 0.

przenies_produkt(X, NowySektor) :-
    retract(produkt(X, Ilosc, _)),
    assertz(produkt(X, Ilosc, NowySektor)).

uzupelnij_stan(X, N) :-
    produkt(X, Ilosc, Sektor),
    NowaIlosc is Ilosc + N,
    retract(produkt(X, Ilosc, Sektor)),
    assertz(produkt(X, NowaIlosc, Sektor)).

    ?- dostepny_produkt(tv).
    Oczekiwany wynik: true
    ?- przenies_produkt(tv, sektor_b).
    Oczekiwany wynik: produkt(tv, 10, sektor_b)
    ?- uzupelnij_stan(pralka, 3).
    Oczekiwany wynik: produkt(pralka, 8, sektor_b)

Zad. 4

% === Definicja miast ===
miasto(warszawa).
miasto(krakow).
miasto(poznan).
miasto(wroclaw).
miasto(gdansk).
miasto(lodz).
miasto(szczecin).
miasto(rzeszow).

% === Połączenia drogowe z czasem przejazdu ===
% droga(Miasto1, Miasto2, Czas)
droga(warszawa, krakow, 4).
droga(krakow, poznan, 3).
droga(warszawa, poznan, 7).
droga(krakow, wroclaw, 2).
droga(poznan, wroclaw, 3).
droga(wroclaw, gdansk, 6).
droga(warszawa, gdansk, 5).
droga(lodz, warszawa, 2).
droga(lodz, wroclaw, 3).
droga(szczecin, poznan, 4).
droga(szczecin, gdansk, 5).
droga(rzeszow, krakow, 3).
droga(rzeszow, warszawa, 6).

% === Połączenie działa w obie strony ===
droga(X, Y, C) :- droga(Y, X, C).

% === Sprawdzenie bezpośredniego połączenia ===
droga(X, Y) :- droga(X, Y, _), !.

% === Określenie czasu przejazdu (uwzględnia trasy pośrednie) ===
czas_przejazdu(X, Y, T) :-
    najkrotsza_trasa(X, Y, _, T), !.

% === Znalezienie NAJKRÓTSZEJ TRASY POD WZGLĘDEM CZASU PRZEJAZDU ===
najkrotsza_trasa(X, Y, Trasa, Czas) :-
    znajdz_najtansza_trase(X, Y, [[X]], 0, Trasa, Czas), !.

% === Znajdowanie najkrótszej trasy rekurencyjnie ===
znajdz_najtansza_trase(X, X, OdwroconaTrasa, Czas, Trasa, Czas) :-
    reverse(OdwroconaTrasa, Trasa), !.

znajdz_najtansza_trase(X, Y, TrasaAktualna, CzasAktualny, NajlepszaTrasa, NajkrotszyCzas) :-
    droga(X, Z, C),
    \+ member(Z, TrasaAktualna), % Unikamy cykli
    NowyCzas is CzasAktualny + C,
    znajdz_najtansza_trase(Z, Y, [Z|TrasaAktualna], NowyCzas, NajlepszaTrasa, NajkrotszyCzas).


    ?- droga(warszawa, krakow).

    Oczekiwany wynik: true

?- czas_przejazdu(warszawa, poznan, T).

    Oczekiwany wynik: T = 7

?- najkrotsza_trasa(warszawa, poznan, X, Y).
    Oczekiwany wynik: X = [[warszawa], krakow, poznan],
                      Y = 7

Zad.5

% === Definicja dynamicznych faktów ===
:- dynamic ksiazka/3.
:- dynamic ocena/3.
:- dynamic preferencje/2.

% === Przykładowe książki ===
ksiazka('W pustyni i w puszczy', sienkiewicz, przygodowa).
ksiazka('Hobbit', tolkien, fantasy).
ksiazka('Pan Tadeusz', mickiewicz, poezja).
ksiazka('Harry Potter', rowling, fantasy).
ksiazka('Zbrodnia i kara', dostojewski, dramat).
ksiazka('Dziady', mickiewicz, dramat).
ksiazka('Władca Pierścieni', tolkien, fantasy).

% === Przykładowe oceny użytkowników ===
ocena(uzytkownik1, 'W pustyni i w puszczy', 5).
ocena(uzytkownik1, 'Hobbit', 4).
ocena(uzytkownik1, 'Pan Tadeusz', 3).
ocena(uzytkownik1, 'Harry Potter', 5).
ocena(uzytkownik1, 'Dziady', 2).
ocena(uzytkownik2, 'Zbrodnia i kara', 5).
ocena(uzytkownik2, 'Dziady', 4).
ocena(uzytkownik2, 'Pan Tadeusz', 4).

% === Preferencje użytkowników dotyczące gatunków ===
preferencje(uzytkownik1, fantasy).
preferencje(uzytkownik2, dramat).

% === Reguła rekomendacji książek ===
polecane_ksiazki(Uzytkownik, Gatunek, ListaKsiazek) :-
    preferencje(Uzytkownik, Gatunek),
    findall(Ksiazka, (ksiazka(Ksiazka, _, Gatunek), ocena(Uzytkownik, Ksiazka, Ocena), Ocena >= 4), ListaKsiazek),
    ListaKsiazek \= [].

   ?- polecane_ksiazki(uzytkownik1, fantasy, Ksiazki).
   Ksiazki = ['Hobbit', 'Harry Potter']
```
